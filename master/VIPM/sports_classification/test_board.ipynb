## White Balance Tests

```python
testImage = Image.open("data/test_degradato/formula 1 racing/4.jpg").convert('RGB')

convert = transforms.PILToTensor()

testTensor = convert(testImage)

print(testTensor.shape)

maxRGB = testTensor.amax(dim=(1,2))

print(maxRGB)

coeff = maxRGB / 255

print(coeff)

wbImage = testTensor * coeff[:, None, None]
```
